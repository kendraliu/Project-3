```python
from sqlalchemy import create_engine, inspect
import pandas as pd
import datetime as dt
import geopandas as gpd
import numpy as np
```

```python
engine = create_engine(f"sqlite:///../Data/InputData/FPA_FOD_20221014.sqlite")
conn = engine.connect()
```

```python
columnNames = inspect(engine).get_columns("Fires")
for column in columnNames:
    print(column["name"], column["type"])
```

```python
#read in filtered database
database = pd.DataFrame(pd.read_sql("SELECT * FROM Fires WHERE STATE == 'CA' AND FIRE_YEAR BETWEEN '2017' AND '2020'" , conn))
database
```

```python
# check to confirm the filtered results
print(database["FIRE_YEAR"].unique())
print(database["STATE"].unique())
print(database[["FIRE_YEAR", "STATE"]])
```

```python
wildfireDatabase = database[["FIPS_NAME", "LATITUDE", "LONGITUDE", "FIRE_NAME", "FIRE_SIZE", "FIRE_SIZE_CLASS", "FIRE_YEAR", "DISCOVERY_DATE", "CONT_DATE", "NWCG_CAUSE_CLASSIFICATION", "NWCG_GENERAL_CAUSE"]]
wildfireDatabase.sample(20)
```

```python
wildfireDatabase = wildfireDatabase.copy()
```

```python
wildfireDatabase["DISCOVERY_DATE"] = pd.to_datetime(wildfireDatabase["DISCOVERY_DATE"], errors="coerce").dt.strftime("%Y%m%d")
wildfireDatabase["CONT_DATE"] = pd.to_datetime(wildfireDatabase["CONT_DATE"], errors="coerce").dt.strftime("%Y%m%d")
wildfireDatabase.dtypes
```

```python
wildfireDatabase
```

```python
wildfireDatabase = wildfireDatabase.dropna()
wildfireDatabase = wildfireDatabase.reset_index(drop="true")
wildfireDatabase
```

```python
wildfireDatabase = wildfireDatabase.rename(columns={"FIPS_NAME": "COUNTY", "NWCG_CAUSE_CLASSIFICATION": "CAUSE_CLASSIFICATION", "NWCG_GENERAL_CAUSE": "CAUSE", "CONT_DATE": "CONTAIN_DATE"})
wildfireDatabase
```

```python
wildfireDatabaseGroupby = wildfireDatabase.groupby("COUNTY").count()
wildfireDatabaseGroupby = wildfireDatabaseGroupby["LATITUDE"]
#wildfireDatabaseGroupby = wildfireDatabaseGroupby.rename(columns={"LATITUDE": "FREQUENCY"})
wildfireDatabaseGroupby = pd.DataFrame(wildfireDatabaseGroupby)
wildfireDatabaseGroupby["FREQUENCY"] = wildfireDatabaseGroupby["LATITUDE"]
wildfireDatabaseGroupby["FREQUENCY"] = wildfireDatabaseGroupby["FREQUENCY"] / wildfireDatabaseGroupby["FREQUENCY"].max()
wildfireDatabaseGroupby = wildfireDatabaseGroupby["FREQUENCY"]
wildfireDatabaseGroupby
```

```python
wildfireDatabase = pd.merge(wildfireDatabase, wildfireDatabaseGroupby, on="COUNTY")
wildfireDatabase
```

# Stats

```python
wildfireCause = wildfireGeoJson.groupby(["CountyName"]).count()
wildfireCause = wildfireCause.reset_index()
wildfireCause = wildfireCause.rename(columns={"LATITUDE": "WILDFIRE_COUNT"})
wildfireCause = wildfireCause[["CountyName", "WILDFIRE_COUNT"]]

wildfireCause
```

```python
causePivot = wildfireGeoJson.copy()

causePivot = causePivot.pivot_table(index='CountyName', columns='CAUSE_CLASSIFICATION', aggfunc='size', fill_value=0)
causePivot.columns = ["CAUSED_BY_HUMAN", "missing_data", "NATURAL_WILDFIRE"]
causePivot["HUMAN/NATURAL"] = causePivot["CAUSED_BY_HUMAN"] / causePivot["NATURAL_WILDFIRE"]
causePivot = causePivot.replace(np.inf, 700)
causePivot["NATURAL/HUMAN"] = causePivot["NATURAL_WILDFIRE"] / causePivot["CAUSED_BY_HUMAN"]
causePivot = causePivot.reset_index()
causePivot = causePivot.drop("missing_data", axis=1)

causePivot
```

```python
wildfirePivot = pd.merge(wildfireCause, causePivot, on="CountyName")
wildfirePivot
```

```python
wildfireSize = pd.DataFrame(wildfireGeoJson.groupby(["CountyName"]).mean()["FIRE_SIZE"])
wildfireSize = wildfireSize.reset_index()
wildfireSize = wildfireSize.rename(columns={"FIRE_SIZE": "AVG_FIRE_SIZE"})
wildfireSize
```

```python
wildfirePivot = pd.merge(wildfirePivot, wildfireSize, on="CountyName")
wildfirePivot
```

```python
mergedGeoJson = CAgeojson.merge(wildfirePivot, on="CountyName")
mergedGeoJson.head()
```

```python
mergedGeoJson.to_file("../Data/OutputData/CaliWildfires.geojson", driver="GeoJSON")
```
